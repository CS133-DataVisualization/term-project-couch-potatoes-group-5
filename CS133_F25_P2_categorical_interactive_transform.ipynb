### Q4: Create an interactive plot
Create an interactive plot using plotly. Your plot can be a previously asked question or a new question, however explain how this interaction strengthen what you want to communicate with your audience.

```python
married_df = og_df[og_df["Marital_Status"].isin(["Married", "Single"])]
married_df.head()
```

```python
married_men_df = married_df[married_df["Gender"] == "Male"]
married_men_df.head()   
```

```python
import plotly.graph_objects as go

ranges = [(0,1), (1,2), (2,3), (3,4), (4,5)] # Define ranges of WFH days (for slider and for loop)
marital_map = {"Single": "lightcoral", "Married": "blue"} # Color map for marital_status

fig = go.Figure()

for i, r in enumerate(ranges):  # One trace per WFH range, colored by marital status
    subset = married_men_df[married_men_df["WFH_Days_Per_Week"].between(r[0], r[1])] # Filter data for the current WFH range 
    
    #https://plotly.com/python/sliders/ 
    fig.add_trace(go.Scatter(
        x=subset["Work_Hours_Per_Week"],
        y=subset["Productivity_Score"],
        mode="markers+text",
        text=subset["Employee_ID"],
        textposition="top center",
        marker=dict(
            size=12,
            color=subset["Marital_Status"].map(marital_map),  # color by marital status
            line=dict(width=1, color="black")
        ),
        name=f"WFH {r[0]}{r[1]} days",
        visible=(i==0)  # only show the first WFH range initially
    ))

steps = [] # Create and add slider steps
for i, r in enumerate(ranges):
    step = dict(
        method="update",
        args=[{"visible": [j==i for j in range(len(ranges))]},
              {"title": f"Work Hours vs Productivity (WFH {r[0]}{r[1]} days) - Married = blue, Single = lightcoral"}],label=f"{r[0]}{r[1]}")
    steps.append(step)

sliders = [dict(
    active=0,
    currentvalue={"prefix": "Work-From-Home Range: "},
    pad={"t": 50},
    steps=steps
)]

# Layout (https://www.geeksforgeeks.org/python/how-to-make-sliders-in-plotly/ )
fig.update_layout(
    title="Work Hours vs Productivity (WFH slider)",
    xaxis=dict(title="Weekly Work Hours"),
    yaxis=dict(title="Productivity Score"),
    sliders=sliders,
    legend_title="WFH Range"
)

fig.show()
```

*The graph above is a [temporary] interactive plotly graph that depicts `Work_Hours_Per_Week` against `Productivity_Score` amongst Married (blue) and Single (light coral) employees. The interactive aspect of this graph is in the form of a slider at the bottom that can control the visualization of data based on `WFH_Days_Per_Week` (Work-From-Home days). Upon, graphing this data, we have found that the data is more dense and higher in productivity scores within the 2-4 work from home days, meaning that is the way most employees seem to be functioning the best regardless of marital status.*
